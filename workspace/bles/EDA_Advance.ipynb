```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

data_path = './data/train/'
```

```python
train_df = pd.read_csv(os.path.join(data_path, 'train_ratings.csv'))
train_df.info()
```

```python
directors_df = pd.read_csv(os.path.join(data_path, 'directors.tsv'), sep='\t')
genres_df = pd.read_csv(os.path.join(data_path, 'genres.tsv'), sep='\t')
titles_df = pd.read_csv(os.path.join(data_path, 'titles.tsv'), sep='\t')
writers_df = pd.read_csv(os.path.join(data_path, 'writers.tsv'), sep='\t')
years_df = pd.read_csv(os.path.join(data_path, 'years.tsv'), sep='\t')
```

### Interaction Top-300인 item에 대한 정보
- Interaction이 많은 item 중에서 특출나게 빈도가 높은 피처는 없는 것 같다.

```python
# interaction이 많은 item 추출 (Top-300)
best_item = list(train_df['item'].value_counts()[:300].index)
```

```python
# interaction이 많은 item 중에서 특출나게 빈도가 높은 감독은 없다.
sns.set(style="whitegrid")
sns.displot(directors_df[directors_df['item'].isin(best_item)]['director'].value_counts())
plt.xlabel("Number of director's interaction ")
plt.ylabel('Number of director')
plt.title('Interaction Top-300 - Directors Frequnecy')
plt.show()
```

```python
# interaction이 많은 item 중에서 특출나게 빈도가 높은 장르는 없다.
sns.set(style="whitegrid")
sns.displot(genres_df[genres_df['item'].isin(best_item)]['genre'].value_counts())
plt.xlabel("Number of genre's interaction ")
plt.ylabel('Number of genre')
plt.title('Interaction Top-300 - Gerens Frequnecy')
plt.show()
```

```python
# interaction이 많은 item 중에서 특출나게 빈도가 높은 작가는 없다.
sns.set(style="whitegrid")
sns.displot(writers_df[writers_df['item'].isin(best_item)]['writer'].value_counts())
plt.xlabel("Number of writer's interaction ")
plt.ylabel('Number of writer')
plt.title('Interaction Top-300 - Writers Frequnecy')
plt.show()
```

```python
# interaction이 많은 item 중에서 특출나게 빈도가 높은 연도는 없다.
sns.set(style="whitegrid")
sns.displot(years_df[years_df['item'].isin(best_item)]['year'].value_counts())
plt.xlabel("Number of year's interaction ")
plt.ylabel('Number of year')
plt.title('Interaction Top-300 - Years Frequnecy')
plt.show()
```
